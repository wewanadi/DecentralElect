```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

name = 'cluster2'
# Read the CSV file into a pandas DataFrame
data = pd.read_csv(f'{name}.csv')

data = data.sort_values(by='Voter.num', ascending=True)

ss = {
    'total_gas_spend': 'Total Gas Spend',
    'manager_gas_spend': 'Initiator Gas Spend',
    'voter_gas_spend': 'Voters Gas Spend',
      }

# Group by the 'Method' column
methods = data.groupby('Cluster_size')

cluster_labels = {
    45: 'PDE45',
    35: 'PDE35',
    25: 'PDE25',
    15: 'PDE15',
    5: 'PDE5',
    0: 'NE',
}

method_colors = {
    0: {'line': '#889C9B', 'scatter': '#889C9B'},
    5: {'line': '#D9AB9A', 'scatter': '#D9AB9A'},
    15: {'line': '#F26444', 'scatter': '#F26444'},
    25: {'line': '#D94436', 'scatter': '#D94436'},
    35: {'line': '#735340', 'scatter': '#735340'},
    45: {'line': '#731D1D', 'scatter': '#731D1D'}
}

# Set style


for spend in ['total_gas_spend','manager_gas_spend','voter_gas_spend']:
    sns.set_style("whitegrid")
    plt.figure(figsize=(5, 3.5))
    plt.rcParams["font.family"] = "Times New Roman"
    # Iterate over each group in reverse order and plot the regression line with appropriate color
    for method, group_data in reversed(list(methods)):
        x = group_data['Voter.num'].tolist()
        a = group_data[spend]/1000000000
        y = a.tolist()
        
        if int(method) not in method_colors: continue
        
        colors = method_colors.get(method, {'line': '#4C72B0', 'scatter': '#55A868'})  # Use default colors if method not found in method_colors
        sns.regplot(x=x, y=y, order=2, color=colors['line'], scatter_kws={'color': colors['scatter'], 's': 30}, line_kws={'linewidth': 2}, label=cluster_labels.get(method, str(method)))

    # Customize the plot
    plt.xlabel('Number of voters', fontsize=16, fontweight='bold')
    plt.ylabel(f'{ss[spend]} (Gwei)', fontsize=15, fontweight='bold')
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.legend(fontsize=12)

    plt.xlim()
    plt.xticks([50, 250, 450, 650, 850]) 
    # Show the plot
    plt.tight_layout()
    os.makedirs('./fig/', exist_ok = True)
    plt.savefig(f'./fig/cluster{spend}.svg', format='svg')
    plt.show()

sns.set_style("whitegrid")
plt.figure(figsize=(5, 3.5))
plt.rcParams["font.family"] = "Times New Roman"
# Iterate over each group in reverse order and plot the regression line with appropriate color
spend = 'voter_gas_spend'
for method, group_data in reversed(list(methods)):
    x = group_data['Voter.num'].tolist()
    a = group_data[spend]/1000000000
    if spend == 'voter_gas_spend':
        a = a / x
    y = a.tolist()
    if int(method) not in method_colors: continue
    
    colors = method_colors.get(method, {'line': '#4C72B0', 'scatter': '#55A868'})  # Use default colors if method not found in method_colors
    sns.regplot(x=x, y=y, order=2, color=colors['line'], scatter_kws={'color': colors['scatter'], 's': 30}, line_kws={'linewidth': 2}, label=cluster_labels.get(method, str(method)))

# Customize the plot
plt.xlabel('Number of voters', fontsize=16, fontweight='bold')
plt.ylabel('Gas Spend per Voter (Gwei)', fontsize=15, fontweight='bold')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=12)

plt.xlim()
plt.xticks([50, 250, 450, 650, 850]) 
# Show the plot
plt.tight_layout()
os.makedirs('./fig/', exist_ok = True)
plt.savefig(f'./fig/cluster_per_voter.svg', format='svg')
plt.show()
```
